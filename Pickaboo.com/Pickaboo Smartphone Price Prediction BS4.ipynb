Now get similar data for every phone:

1. extract a tags and retrieve href links from it. 
2. Save it as a list

```python
# link_ = "https://www.pickaboo.com/smartphone.html?p=1&product_list_limit=80&product_list_order=price&product_list_dir=desc"

link_ = "https://www.pickaboo.com/smartphone.html?p=2&product_list_limit=80&product_list_order=price&product_list_dir=desc"

r_link = requests.get(link_)
soup_link = bs(r_link.content)
```

```python
tag_links = soup_link.find_all(class_ = "product name product-item-name")
```

Get Spec information for all the phones in Pickaboo site

```python
#function to get phone information.
def get_phone_info(link):
    soup = bs(requests.get(link).content)

    product_attibute = soup.find(class_ = "data table additional-attributes")
    product_rows = product_attibute.find_all("tr")

    phone_info = {}
    for i in product_rows:
        content_key = i.find("th").get_text()
        content_value = i.find("td").get_text()
        phone_info[content_key] = content_value
    
    return phone_info
```

```python
list_of_phone_info = []
```

Saving all the dictionary values in a list

```python

for i in range(len(tag_links)):
    try:
        individual_link = tag_links[i].find(class_ = "product-item-link")['href']
        list_of_phone_info.append(get_phone_info(individual_link))
    except AttributeError as a:
        pass
```

Saving all the dictionary information in JSON file

```python
import json

def save_data(title, data):
    with open(title, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)
```

```python
import json

def load_data(title):
    with open(title, 'w', encoding='utf-8') as f:
        return json.load(f)
```

```python
save_data("list_of_phone_info.json", list_of_phone_info)
```

Let's add the price 

```python
# price_link = "https://www.pickaboo.com/smartphone.html?p=1&product_list_limit=80&product_list_order=price&product_list_dir=desc"
price_link = "https://www.pickaboo.com/smartphone.html?p=2&product_list_limit=80&product_list_order=price&product_list_dir=desc"

soup_price = bs(requests.get(price_link).content)
```
